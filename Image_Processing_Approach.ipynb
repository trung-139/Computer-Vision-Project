```python
import cv2
import numpy as np
import numpy as np
from skimage.transform import hough_line, hough_line_peaks
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('263205_sat.jpg')
```

```python
# Convert to grayscale
gray_1 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply thresholding (adjust threshold value as needed)
thresh_1 = cv2.threshold(gray_1, 95, 255, cv2.THRESH_BINARY)[1]

# Apply Canny edge detection
edges_1 = cv2.Canny(thresh_1, 50, 150)

# Apply morphological closing (optional, adjust kernel size as needed)
kernel_1 = np.ones((10, 10), np.uint8)
closed_1 = cv2.morphologyEx(edges_1, cv2.MORPH_CLOSE, kernel_1)

# Create a figure with 2x2 subplots (adjust layout as needed)
fig, axes = plt.subplots(2, 2, figsize=(10, 6))  # Customize figure size

# Display original image (top-left)
axes[0, 0].imshow(image, cmap='gray')
axes[0, 0].set_title('Original Image')
axes[0, 0].axis('off')  # Remove unnecessary ticks and labels

# Display binary threshold image (top-right)
axes[0, 1].imshow(thresh_1, cmap='gray')
axes[0, 1].set_title('Binary Threshold')
axes[0, 1].axis('off')

# Display edge-detected image (bottom-left)
axes[1, 0].imshow(edges_1, cmap='gray')
axes[1, 0].set_title('Edges')
axes[1, 0].axis('off')

# Display closed image (bottom-right)
axes[1, 1].imshow(closed_1, cmap='gray')
axes[1, 1].set_title('Closed Image')
axes[1, 1].axis('off')

# Adjust layout (optional)
fig.tight_layout()

# Display the visualizations
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

msk = cv2.imread('263205_mask.png')

# Apply Canny edge detection (adjust parameters as needed)
edges_1 = cv2.Canny(image, 50, 150)

# Apply morphological closing (optional, adjust kernel size as needed)
kernel_1 = np.ones((10, 10), np.uint8)
closed_1 = cv2.morphologyEx(edges_1, cv2.MORPH_CLOSE, kernel_1)

# Create a figure with 1 row and 3 columns
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # Customize figure size

# Display original image (left)
axes[0].imshow(image)  # Use 'bgr' for original color image
axes[0].set_title('Original Image')
axes[0].axis('off')

# Display mask image (center)
axes[1].imshow(msk, cmap='gray')
axes[1].set_title('Mask Image')
axes[1].axis('off')

# Display closed image with edge detection (right)
axes[2].imshow(closed_1, cmap='gray')
axes[2].set_title('Closed Image')
axes[2].axis('off')

# Adjust layout (optional)
fig.tight_layout()

# Display the visualizations
plt.show()
```
